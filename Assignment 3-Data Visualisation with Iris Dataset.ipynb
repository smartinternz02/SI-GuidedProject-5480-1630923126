```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
iris=pd.read_csv('Iris.csv')
iris.head()
```

```python
iris
```

```python
iris["Species"].value_counts()
```

```python
iris.plot(kind="scatter", x="SepalLengthCm", y="SepalWidthCm")
plt.show()
```

```python

iris.plot(kind="scatter", x="SepalLengthCm", y="SepalWidthCm",color="green",s=70 )
plt.show()
```

```python
sns.jointplot(x="SepalLengthCm", y="SepalWidthCm", data=iris, size=5)
plt.show()
```

```python

sns.FacetGrid(iris, hue="Species", size=5) \
   .map(plt.scatter, "SepalLengthCm", "SepalWidthCm") \
   .add_legend()
plt.show()
```

```python
KS = {'color': ['blue', 'red', 'yellow']}
sns.FacetGrid(iris, hue_kws=KS, hue="Species", size=5) \
   .map(plt.scatter, "SepalLengthCm", "SepalWidthCm") \
   .add_legend()
plt.show()
```

```python
sns.boxplot(x="Species", y="PetalLengthCm", data=iris )
plt.show()
```

```python
ax= sns.boxplot(x="Species", y="PetalLengthCm", data=iris)
ax= sns.stripplot(x="Species", y="PetalLengthCm", data=iris, jitter=True, edgecolor="gray")
plt.show()
```

```python
sns.violinplot(x="Species", y="PetalLengthCm", data=iris, size=6)
plt.show()
```

```python
sns.FacetGrid(iris, hue="Species", size=5) \
   .map(sns.kdeplot, "PetalLengthCm") \
   .add_legend()
plt.show()
```

```python
sns.pairplot(iris.drop("Id", axis=1), hue="Species", size=2)
plt.show()
```

```python
sns.pairplot(iris.drop("Id", axis=1), hue="Species", size=2, diag_kind="kde")
plt.show()
```
